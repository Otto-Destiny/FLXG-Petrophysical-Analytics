# Log Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import lasio
```

```python
well = pd.read_csv("Neptune_data.csv", header=0)
well.replace(-999.25, np.nan, inplace=True)
well.describe()
```

```python
fig=plt.subplots(figsize=(20,20))

ax1 =plt.subplot2grid((1,3),(0,0), rowspan=2,colspan=1)
ax2 =plt.subplot2grid((1,3),(0,1))
ax3 =plt.subplot2grid((1,3),(0,2))

ax1.plot("GR", 'DEPTH',data=well,color="yellow")
ax1.set_title("Gamma Log")
ax1.set_xlim(0,200)
ax1.grid()

ax2.plot("RHOB", 'DEPTH',data=well,color="red")
ax2.set_title("Density Log")
#ax2.set_xlim(0,200)
ax2.grid()

ax3.plot("NPHI", 'DEPTH',data=well,color="blue")
ax3.set_title("Neutron Log")
#ax3.set_xlim(0,200)
ax3.grid()
```

```python
fig=plt.subplots(figsize=(10,10))

ax1 =plt.subplot2grid((1,3),(0,0))
ax2 =plt.subplot2grid((1,3),(0,1))
ax3 =plt.subplot2grid((1,3),(0,2))
ax4 =ax2.twiny()

ax5=ax1.twiny()
ax5.xaxis.set_visible(False)
ax6=ax2.twiny()
ax6.xaxis.set_visible(False)
ax7=ax3.twiny()
ax7.xaxis.set_visible(False)

#Gamma Track
ax1.plot("GR","DEPTH",color="green",data=well)
ax1.set_xlabel("Gamma")
ax1.xaxis.label.set_color("green")
ax1.set_xlim(0,200)
ax1.set_ylabel("Depth")
ax1.tick_params(axis='x',colors='green')
ax1.spines['top'].set_edgecolor('green')
ax1.set_xticks([0,50,100,150,200])


#Density track
ax2.plot("RHOB","DEPTH",color="red",data=well)
ax2.set_xlabel("Density")
ax2.xaxis.label.set_color("red")
ax2.set_xlim(2200,2800)
#ax2.set_ylabel("Depth")
ax2.tick_params(axis='x',colors='red')
ax2.spines['top'].set_edgecolor('red')
#ax2.set_xticks([0,50,100,150,200])

#Sonic Track
ax3.plot("DRHO","DEPTH",color="yellow",data=well)
ax3.set_xlabel("Sonic")
ax3.xaxis.label.set_color("yellow")
ax3.set_xlim(0,100)
#ax3.set_ylabel("Depth")
ax3.tick_params(axis='x',colors='yellow')
ax3.spines['top'].set_edgecolor('yellow')
ax3.set_xticks([0,50,100,150,200])

#Neutron on top of density
ax4.plot("NPHI","DEPTH",color="blue",data=well)
ax4.set_xlabel("Neutron")
ax4.xaxis.label.set_color("blue")
ax4.set_xlim(0.01,0.5)
ax4.set_ylabel("Depth")
ax4.tick_params(axis='x',colors='blue')
ax4.spines['top'].set_position(("axes", 1.08))
ax4.spines['top'].set_visible(True)
ax4.spines['top'].set_edgecolor('blue')
#ax4.set_xticks([0,50,100,150,200])

for ax in [ax1, ax2, ax3]:
    ax.grid(which='major', color='lightgrey',linestyle='-')
    ax.xaxis.set_ticks_position('top')
    ax.xaxis.set_label_position('top')
    ax.spines['top'].set_position(('axes', 1.02))

plt.tight_layout()

    




```

```python
las = lasio.read('well3.LAS')
df = las.df()
df['DEPTH'] = df.index
```

```python
df
```

```python
fig = plt.subplots(figsize=(7,10))

ax1 = plt.subplot2grid((1,1), (0,0), rowspan=1, colspan=1)
ax2 = ax1.twiny()

ax1.plot('DEN', 'DEPTH', data=df, color='red', lw=0.5)
ax1.set_xlim(1.95, 2.95)
ax1.set_xlabel('Density')
ax1.xaxis.label.set_color("red")
ax1.tick_params(axis='x', colors="red")
ax1.spines["top"].set_edgecolor("red")

ax2.plot('NEU', 'DEPTH', data=df, color='blue', lw=0.5)
ax2.set_xlim(45, -15)
ax2.set_xlabel('Neutron')
ax2.xaxis.label.set_color("blue")
ax2.spines["top"].set_position(("axes", 1.08))
ax2.tick_params(axis='x', colors="blue")
ax2.spines["top"].set_edgecolor("blue")

x1=df['DEN']
x2=df['NEU']

x = np.array(ax1.get_xlim())
z = np.array(ax2.get_xlim())

nz=((x2-np.max(z))/(np.min(z)-np.max(z)))*(np.max(x)-np.min(x))+np.min(x)

ax1.fill_betweenx(df['DEPTH'], x1, nz, where=x1>=nz, interpolate=True, color='green')
ax1.fill_betweenx(df['DEPTH'], x1, nz, where=x1<=nz, interpolate=True, color='yellow')

for ax in [ax1, ax2]:
    ax.set_ylim(4400, 4300)
    ax.xaxis.set_ticks_position("top")
    ax.xaxis.set_label_position("top")
    
```
