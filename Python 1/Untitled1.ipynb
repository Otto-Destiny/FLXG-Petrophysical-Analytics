```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
a = np.linspace(0, 360, 50)
arad = np.deg2rad(a)
```

```python
y1 = np.sin(arad)
y2 = np.cos(arad)
y2
```

```python
###plot this function
plt.plot(arad, y1)
plt.plot(arad, y2)
plt.title("Sine and cosine plot", size =30)
plt.show()
```

```python
frame =pd.DataFrame({'x':a, 'sin x':y1, 'cos x':y2})
frame
```

```python
y3 = np.tan(arad)
frame["tanx"] =y3
frame
```
