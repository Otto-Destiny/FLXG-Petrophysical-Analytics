## Shale Volume Calculations

```python
def VshFun(GR_log, GR_min, GR_max, correction=None):

    Ish =(GR_log-GR_min)/(GR_max-GR_min)       #Linear Gamma Ray
    larionov_young=0.083*(2**(3.7*Ish)-1)   #Larionov (1969) - Tertiary rocks
    larionov_old=0.33*(2**(2*Ish)-1)        #Larionov (1969) - Older rocks
    clavier=1.7-(3.38-(Ish+0.7)**2)**0.5    #Clavier (1971)
    steiber=0.5*Ish/(1.5-Ish)               #Steiber (1969) - Tertiary rocks

    if correction == "young":
        Vsh = larionov_young
    elif correction == "older":
        Vsh = larionov_old
    elif correction=="clavier":
        Vsh = clavier
    elif correction=="steiber":
        Vsh = steiber
    else:
        Vsh = Ish
    return Vsh
```

```python
# Looping through the section

Sec['Vsh'] = VshFun(Sec.GR, df.GR.min(), df.GR.max())
```

## Porososity Calculations

```python
#Density

def phid_sh_corr(den, den_ma, den_fl, den_sh, Vsh):
    phid = (den - den_ma) / (den_fl - den_ma)
    phid_sh = (den_sh - den_ma) / (den_fl - den_ma)
    phid_sh_corr = phid - Vsh * phid_sh
    return phid_sh_corr

# NEUTRON:
def phin_sh_corr(neut, neut_sh, Vsh):
    phin_sh_corr = (neut-Vsh*neut_sh)/100
    return phin_sh_corr

#Neutron-Density
def phixnd(phinshc, phidshc):
    phixnd= (phinshc + phidshc) / 2
    return phixnd
```

```python
# Matrix, fluid and shales values
den_ma, den_fl, den_sh = 2.65, 1.1, 2.4
SH_index = df.GR.idxmax()
phin_sh = df.NPHI[SH_index]
# Looping through the section
PHID = phid_sh_corr(Sec.RHOB, den_ma, den_fl, den_sh, Sec.Vsh).clip(0,1)
PHIN = phin_sh_corr(Sec.NPHI, phin_sh, Sec.Vsh).clip(0,1)

Sec['PHI'] = phixnd(PHIN, PHID).clip(0,1)
```

## Water Saturation Calculations

```python
a= 1 #turtuosity factor
m= 1.6 #cementation factor
n= 1.8 #saturation exponent
Rw = 0.087 # water resitivity
```

```python
def archie(Rw, Rt, PHI, a, m, n):
        F = a / (PHI**m)
        archie = (F * Rw/Rt)**(1/n)
        return archie
```

```python
Sec['SW'] = archie(Rw, Sec.ILD, Sec.PHI,a,m,n).clip(0,1)
Sec['Matrix'] = 1 - Sec.Vsh - Sec.PHI
```
